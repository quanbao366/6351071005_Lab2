```python
# Hàm tính F-measure
def f_measure(precision, recall):
    if precision + recall == 0:
        return 0
    return 2 * precision * recall / (precision + recall)

# Precision và Recall của nhãn >50K
precision_values = [0.64, 0.65, 0.28]  # ID3, CART, Naive Bayes
recall_values = [0.60, 0.62, 0.99]
print("Trần Quân Bảo - 6351071005")
# Tính F-measure cho từng mô hình
f_measures = []
for p, r in zip(precision_values, recall_values):
    f = f_measure(p, r)
    f_measures.append(f)
    print(f"Precision: {p}, Recall: {r}, F-measure: {f:.4f}")

```
